```python
#Opcion 4: Imputacion simple
#Imputar con un valor puntual
df['COMM'] = df['COMM'].fillna(value=0)
df
```

```python
#Imputacion a partir de la media
promedio = df['WEIGHT'].mean(skipna = True)
```

```python
promedio
```

```python
df['WEIGHT'] = df['WEIGHT'].fillna(value=promedio)
df
```

```python
df['HOBBY'] = df['HOBBY'].fillna(value='NO HOBBY')
df
```

```python
mapeo_hobby = {'-' : 'NO HOBBY'}
```

```python
df['HOBBY'] = df['HOBBY'].replace(mapeo_hobby)
df
```

```python
df = df.rename(columns={'SAL':'SALARIO', 'COMM':'COMISION', 'WEIGHT':'PESO','ENAME':'NOMBRE', 'JOB':'CARGO', 'HIREDATE':'FECHA_INGRESO', 'HOBBY':'PASATIEMPO', 'EDUCATION':'EDUCACION', 'MARITAL_STATUS':'ESTADO_CIVIL'})
df
```

```python
#MANEJO DE DATOS ATIPICOS
#Calcular Q1, Q3, MEDIANA, IQR, Li, Ls
Q1 = df['PESO'].quantile(0.25)
Q3 = df['PESO'].quantile(0.75)
MEDIANA = df['PESO'].median()
IQR = Q3 - Q1
Li = Q1 - 1.5 * IQR
Ls = Q3 + 1.5 * IQR
print(Q1, Q3, MEDIANA, IQR, Li, Ls)
```

```python
from IPython.display import Markdown as md
```

```python
md("### Q1 es igual a {}, Q3 es igual a {}, IQR es igual a {}, limite inferior es igual a {}, limite superior es igual a {} ".format(Q1, Q3, IQR, Li, Ls))
```

```python
#Graficar botsplot
flierprops=dict(marker='o', markerfacecolor='green', markersize=7, linestyle='none')
medianprops=dict(linestyle='-', linewidth=2.5, color='red')
sns.boxplot(x=df['PESO'], fliersize=5, fill=True, flierprops=flierprops, medianprops=medianprops)
plt.title('Boxplot de la variable PESO')
```

```python
df['PESO'] = df['PESO'].map(lambda x: Li if x < Li else Ls if x > Ls else x)
```

```python
df
```

# VISUALIZAR COLUMNA CARGO

Value_conts me permite obtener un conteo de valores unicos.

```python
grupos_cargo = df['CARGO'].value_counts().sort_index()
grupos_cargo.plot(kind='bar', color=sns.color_palette("pastel"))
plt.title('Grafico de barras de la variable CARGO')
plt.xlabel('CARGO')
plt.ylabel('Cantidad')
plt.show()
```

```python
#Version 2 con agrupacion
df.groupby('CARGO').size().plot(kind='bar', color=sns.color_palette("pastel"))
plt.title('Grafico de barras de la variable CARGO')
plt.xlabel('CARGO')
plt.ylabel('Cantidad')
plt.show()
```

```python
#Graficar en pastel la columna pasatiempo con porcentaje
df['PASATIEMPO'].value_counts().plot(kind='pie', autopct='%0.1f%%')
plt.title('Grafico de pastel de la variable PASATIEMPO')
plt.show()
```

```python
#Graficar la relacion entre cargo y estado civil
conteo_cargo_estado_civil = df.groupby(['CARGO', 'ESTADO_CIVIL']).size().unstack(fill_value = 0)
conteo_cargo_estado_civil
```

```python
conteo_cargo_estado_civil.plot(kind='bar',figsize=(5,5), color=sns.color_palette("pastel"))
plt.title('Grafico de barras de la relacion entre CARGO y ESTADO_CIVIL')
plt.xlabel('Cargo')
plt.ylabel('Cantidad')
plt.show()
```

```python
#Codificacion one-hot encoder
#Transformacion usando pandas get_dummies()
df = pd.get_dummies(df, columns=['ESTADO_CIVIL'], dtype = int, drop_first=True)
df
```

```python
#Transformacion usando sklearn
from sklearn.preprocessing import OneHotEncoder
```

```python
#Crear un encoder
encoder = OneHotEncoder()
#Se hace un fitting del encoder es decir un ajuste o se adapta el encoder a los datos
data = encoder.fit_transform(df[['EDUCACION']]).toarray()
print(data)
```

```python
#Conocer las categorias
encoder.categories_
```

```python
#Opcion 1: crear un dataframe con esa columnas con valores enteros
df_one_hot = pd.DataFrame(data, columns=encoder.categories_[0])
df_one_hot.astype(int)
```

```python
#Concatenar el df de codificacion de la columna educacion al df original eliminando
df = pd.concat([df, df_one_hot], axis=1)
df
```

```python
#Borrar columna educacion
df = df.drop(columns=['EDUCACION',"BACHELOR'S DEGREE"])
df
```

```python
df['HIGH SCHOOL'] = df['HIGH SCHOOL'].astype(int)
df["MASTER'S DEGREE"] = df["MASTER'S DEGREE"].astype(int)
df
```

```python
#Consultar el tipo de dato de la Columna fecha de ingreso
df['FECHA_INGRESO'].dtype
```

```python
#Convertir la columna fecha ingreso a datetime
df['FECHA_INGRESO'] = pd.to_datetime(df['FECHA_INGRESO'], dayfirst=True, errors='coerce')
```

```python
#Generar las columna de MES y ANIO
df['MES'] = df['FECHA_INGRESO'].dt.month
df['ANIO'] = df['FECHA_INGRESO'].dt.year
```
